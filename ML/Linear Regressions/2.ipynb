```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("height-weight.csv")
df.head()
```

```python
X = df["Weight"].values
y = df["Height"].values
```

```python
X_mean = X.mean()
y_mean = y.mean()
```

```python
coef = (((X-X_mean)*(y-y_mean)).sum()/((X-X_mean)**2).sum()).round()
```

```python
coef
```

```python
inter = (y_mean-coef*X_mean).round()
```

```python
inter
```

```python
def predict(coef , inter , x):
    return coef*x+inter
```

```python
predict(coef , inter , 100).round()
```
