```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/mishravipul/data/raw/main/simple_linear_data.csv")
```

```python
df.insert(0 , "bias" , 1) 
```

```python
df.head()
```

```python
X = df[["bias" , "G2"]].values
y = df["G3"].values
```

```python
def hypothesis(X , wieght):
    return X @ wieght
```

```python
XT = X.T
w = np.linalg.inv(XT @ X) @XT @ y
```

```python
def MSE(y , hypothesis):
    mse = np.mean(np.square(y - hypothesis))
    return mse
```

```python
MSE(y , hypothesis(X , w))
```

```python
def gradient(X , hypothesis , y):
    n = len(y)
    gradient = (X.T @ (hypothesis - y))/n
    return gradient
    
```

```python
gradient(X , hypothesis(X , w) , y)
```

```python
def gradient_descent(X , y , w , alpha):
    for _ in range(1000):
        G = gradient(X , hypothesis(X , w) , y)
        w = w - alpha*G

    return w
```

```python
gradient_descent(X , y , w , alpha=0.00001)
```

```python
#removing outlier

mean = df["G3"].mean()
std = df["G3"].std()

data_clean = df[(df["G3"] >= mean - 2*std) & (df["G3"] <= mean + 2*std)]
```

```python
X_clean = data_clean[["bias" , "G2"]].values
y_clean = data_clean["G3"].values
```

```python
X_cleanT = X_clean.T
w_clean = np.linalg.inv(X_cleanT @ X_clean) @ X_cleanT @ y_clean
```

```python
MSE(y_clean , hypothesis(X_clean , w_clean))
```

```python
gradient(X_clean , hypothesis(X_clean , w_clean) , y_clean)
```

```python
gradient_descent(X_clean , y_clean , w_clean , alpha=0.00001)
```
